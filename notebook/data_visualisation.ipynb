```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import math
```

```python
df = pd.read_csv("cleaned_bmw_data.csv")
```

```python
# Create the 'car_age' feature from the 'year' column
# We assume the analysis is being done in 2025
df['car_age'] = 2025 - df['year']

# It's good practice to drop the original 'year' column to avoid redundancy
df.drop('year', axis=1, inplace=True)

print("Created 'car_age' feature and dropped 'year' column.")
print(df[['car_age', 'price']].head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=50)
plt.title('Distribution of BMW Car Prices', fontsize=16)
plt.xlabel('Price (£)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price vs. Car Age & Mileage
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Price vs. Car Age
sns.scatterplot(x='car_age', y='price', data=df, alpha=0.5, ax=axes[0])
axes[0].set_title('Price vs. Car Age')

# Price vs. Mileage
sns.scatterplot(x='mileage', y='price', data=df, alpha=0.5, ax=axes[1])
axes[1].set_title('Price vs. Mileage')

plt.tight_layout()
plt.show()
```

```python
#  Price vs. Transmission & Fuel Type

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Price vs. Transmission
sns.boxplot(x='transmission', y='price', data=df, ax=axes[0])
axes[0].set_title('Price vs. Transmission Type')

# Price vs. Fuel Type
sns.boxplot(x='fuelType', y='price', data=df, ax=axes[1])
axes[1].set_title('Price vs. Fuel Type')

plt.tight_layout()
plt.show()
```

```python
# Select only numerical columns for the correlation matrix
numerical_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
